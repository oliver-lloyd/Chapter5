```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gcn_baseline = pd.read_csv('LOOCV_cosines.csv')[['left_out_node', 'cosine_to_actual']]
gcn_baseline.columns = ['drug', 'cosine_sim']
gcn_baseline['vector_method'] = 'GCN baseline'
gcn_baseline['neighbourhood_method'] = 'drug similarity'

chap4 = pd.read_csv('../../../../../Chapter4/data/results/cosines_vs_actual.csv')
df = pd.concat([chap4, gcn_baseline])
```

```python
plt.clf()
sns.boxplot(data=df, y='cosine_sim', x='vector_method', hue='neighbourhood_method')
plt.title('Similarity of vector estimations to real embeddings')
plt.ylabel('Cosine similarity')
plt.xlabel('Aggregation method')
plt.legend(title='Neighbourhood definition', loc='lower right')
plt.xticks(size=8, rotation=22.5)
plt.tight_layout()
#plt.savefig('fig/cosine_by_agg_and_hood.png')
```
